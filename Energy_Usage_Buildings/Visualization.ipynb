```python
classroom = pd.read_csv(file_path + 'classroom_cleaned.csv')

classroom['Date'] = pd.to_datetime(classroom['Date'])
classroom = classroom.sort_values(by = 'Date')
classroom = classroom.reset_index(drop = True)
```

```python
classroom.head()
```

```python
layout = dict(
    title = 'Classroom Energy Usage Statistics',
    xaxis = dict(
        rangeselector = dict(
            buttons = list([
                dict(count = 60,
                    label = '5 year',
                    step = 'month',
                    stepmode = 'backward'),
                dict(count = 36,
                    label = '3 year',
                    step = 'month',
                    stepmode = 'backward'),
                dict(count = 12,
                    label = '1 year',
                    step = 'month',
                    stepmode = 'backward'),
                dict(step = 'all')
            ])
        ),
        
        rangeslider = dict(
            visible = True
        ),
        
        type = 'date'
    ),
    
    yaxis = dict(
        title = 'kBtu'
    )
)

trace1 = go.Scatter(
    x = classroom['Date'],
    y = classroom['Chilled Water'],
    name = 'Classroom Chilled Water'
)

trace2 = go.Scatter(
    x = classroom['Date'],
    y = classroom['Electricity'],
    name = 'Classroom Electricity'
)

trace3 = go.Scatter(
    x = classroom['Date'],
    y = classroom['Steam'],
    name = 'Classroom Steam'
)

fig = go.Figure(data = [trace1, trace2, trace3], layout = layout)
iplot(fig)
```
